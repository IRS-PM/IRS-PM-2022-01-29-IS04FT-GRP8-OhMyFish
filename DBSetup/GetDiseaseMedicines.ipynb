### Get list of medicines from Disease dataset

```python
## import libraries
import pandas as pd
```

```python
# load dataset
filename = 'FishDiseases(clean) - Disease.csv'
folderpath = r'/home/jengwei/nus/1. Intelligent Reasoning Systems/1. Machine Reasoning/MVP Project/Dataset//'

disease_df = pd.read_csv(folderpath + filename, encoding = "ISO-8859-1", index_col='ID')
disease_df.head()
```

```python
# get list of medicines
medicines = []
start_col_idx = disease_df.columns.get_loc('Medicine1')
end_col_idx = disease_df.columns.get_loc('Medicine8')

for i in range(disease_df.shape[0]):
    for j in range(start_col_idx, end_col_idx):
        if (pd.isnull(disease_df.iloc[i,j]) == False) and (disease_df.iloc[i,j] not in medicines):
            medicines.append(disease_df.iloc[i,j])

print(medicines)
df = pd.DataFrame(medicines, columns = ['Medicine'])
df.head()
```

```python
filename = 'MedicineList.csv'
folderpath = r'/home/jengwei/nus/1. Intelligent Reasoning Systems/1. Machine Reasoning/MVP Project/Dataset//'

df.to_csv(folderpath+filename)
```
